```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = joblib.load("lungcancer_svm.pkl")
scaler = joblib.load("lungcancer_scaler.pkl")
```

```python
X_test = pd.read_csv("lung_cancer_X_test.csv")
y_test = pd.read_csv("lung_cancer_y_test.csv")
```

```python
X_test_scaled = scaler.transform(X_test)

predictions = model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Test Accuracy: {accuracy:.4f}")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", report)

output = pd.DataFrame({"LUNG_CANCER_PREDICTION": predictions})
output.to_csv("lung_cancer_predictions.csv", index=False)
print("Predictions saved to lung_cancer_predictions.csv")
```
